### 객체 검출

* 하나의 이미지 검출

```python
img_file = '사진/cat.jfif'
img_type = img_file.split('.')[-1]
img_type
```

```python
# JFIF파일은 JPG로 변경
img_type = 'jpg' if img_type == 'jfif' else img_type
```

```python
import urllib3
import json
import base64
openApiURL = "http://aiopen.etri.re.kr:8000/ObjectDetect"
```

```python
with open(img_file, 'rb') as file:
    img_contents = base64.b64encode(file.read()).decode("utf8")
```

```python
request_json = {
    "access_key": ai_key,
    "argument": {
        "type": img_type,
        "file": img_contents
    }
}
```

```python
http = urllib3.PoolManager()
response = http.request(
    "POST",
    openApiURL,
    headers={"Content-Type": "application/json; charset=UTF-8"},
    body=json.dumps(request_json)
)
```

```python
response.status
```

```python
response.data
```

```python
result = json.loads(response.data)
result
```

```python
len(ai_key)
```

```python
obj_list = result['return_object']['data']
obj_list[0]
```

- 시각화

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
img = Image.open(img_file)
plt.imshow(img);
```

```python
draw = ImageDraw.Draw(img)
obj = obj_list[0]
name = obj['class']
x = int(obj['x'])
y = int(obj['y'])
w = int(obj['width'])
h = int(obj['height'])
draw.rectangle(((x, y), (x+w, y+h)), outline = (255, 0, 0), width = 1)  # 두 지점을 튜플로 묶어야 한다
draw.text((x+10, y+3), name, font = ImageFont.truetype('malgun.ttf', 15), fill = (255, 0, 0))
```
