# Simple RNN(Many-to-one 문제)
- sin파형 시계열을 입력하고 다음 스텝의 출력을 예측
- 3개의 순서열을 입력한 후 출력값이 target과 일치하게 만든다

- 데이터 준비

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = np.sin(np.arange(21) * np.pi/4)         # pi / 4는 45도
s
```

```python
plt.plot(s, 'ro-')
plt.xlim(-0.5, 20.5)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.show()
```

- X_train, y_train 데이터 만들기

```python
# 첫 번째에는 1, 2, 3번째 점이 X가 되고 4번째 점이 y가 된다
# 마지막에는 18, 19, 20번째 점이 X가 되고 마지막 점이 y가 된다

X_train = []
y_train = np.empty(len(s) - 3)
for i in range(len(s)-3):
    X_train.append(s[i:i+3])
    y_train[i] = s[i+3]
y_train
```

```python
X_train = np.array(X_train).reshape(-1, 3, 1)
X_train.shape   # 18개의 데이터,  3개의 시점(timesteps), input dim = 1차원
```

```python
plt.figure(figsize = (10, 6))
plt.subplot(211)
plt.plot([0, 1, 2], X_train[0].flatten(), 'bo-', label = 'input sequence', )
plt.plot([3], y_train[0], 'ro', label = 'target')
plt.xlim(-0.5, 4.5)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.legend()
plt.title('First sequence')

plt.subplot(212)
plt.plot([1, 2, 3], X_train[1].flatten(), 'bo-', label = 'input sequence')
plt.plot([4], y_train[1], 'ro', label = 'target')
plt.xlim(-0.5, 4.5)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.legend()
plt.title('Second sequence')

plt.show()
```

- 순환 신경망 모델 정의/설정

```python
import tensorflow as tf
seed = 2022
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
np.__version__          # version이 1.20.2에서는 아래 모델 정의가 안된다
```

```python
model = Sequential([
                    SimpleRNN(10, input_shape = (3, 1), activation = 'tanh'),      # input_shape = X_train.shape[1:]
                    Dense(1, activation = 'linear')
])
model.summary()

# input차원이 1차원, 1*10
# node갯수가 10개, 10*10
# bias 10개 10
# 합치면 120
```

```python
model.compile(optimizer = 'adam', loss= 'mse')
```

- 모델 학습 전

```python
plt.plot(np.arange(3, 21), y_train, 'ro-', label = 'target')
plt.plot(np.arange(3, 21), model.predict(X_train), 'bs-', label = 'predict')
plt.xlim(-0.5, 20.5,)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.legend()
plt.title('Before training')
plt.show()
```

- 모델 학습

```python
hist = model.fit(X_train, y_train, epochs = 100, verbose = 0)
```

```python
plt.plot(np.arange(3, 21), y_train, 'ro-', label = 'target')
plt.plot(np.arange(3, 21), model.predict(X_train), 'bs-', label = 'predict')
plt.xlim(-0.5, 20.5,)
plt.ylim(-1.1, 1.1)
plt.grid()
plt.legend()
plt.title('After training')
plt.show()
```

```python
plt.plot(hist.history['loss'])
plt.title('Loss')
plt.grid()
plt.show()
```
